### Quels sont les descripteurs (colonnes) du dataset?

```python
import pandas as pd
import numpy as np


%matplotlib inline
df = pd.read_csv('train_users_2.csv', index_col=0)
display(df)

print("Les descripteurs du dataset")
#df.dtypes
print(df.columns.tolist())


```

### Combien d'enregistrements (lignes) ont été fournis ?

```python
nombre_enregistrements = df.shape[0]
print("le nombre d'enregistrements:",nombre_enregistrements)
```

### Quel est le format des données. Par exemple, dans quel format les dates sont fournies, existe-t-il des valeurs numériques, à quoi ressemblent les différentes valeurs catégorielles ?

```python
types_donnees = df.dtypes

dates = df[['date_account_created', 'timestamp_first_active', 'date_first_booking']]
donnees_numeriques = df[['age', 'signup_flow']]
donnees_categoriels = df[['gender', 'signup_method', 'language', 'affiliate_channel', 'affiliate_provider', 'first_affiliate_tracked', 'signup_app', 'first_device_type', 'first_browser', 'country_destination']]

# Afficher les résultats
print("Types de données :")
print(types_donnees)

print("\nDonnées pour les colonnes de type date :")
display(dates)

print("\nDonnées pour les colonnes numériques :")
display(donnees_numeriques)

print("\nDonnées pour les colonnes catégorielles :")
display(donnees_categoriels)

```

<span>Les colonnes 'date_account_created' et 'date_first_booking' sont au format texte (object) tandis que la colonne 'timestamp_first_active' est un entier (int64)
</span>

### Y a-t-il des valeurs manquantes?

```python
valeurs_manquantes = df.isnull().sum()

display(valeurs_manquantes[valeurs_manquantes > 0])
```

### Est-ce qu’il y’a des dépendances évidentes au niveau des descripteurs?

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_numeric = df.select_dtypes(include=[np.number])

correlation_numeriques = df_numeric.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_numeriques, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de corrélation des variables numériques')
plt.show()
```

<span>Il n'y a pas de corrélation entre les variables numériques</span>
